```python
import serial
import time

def read_serial_data(port, baud_rate, duration, output_file):
    try:
        # Open the serial port
        ser = serial.Serial(port, baud_rate, timeout=1)
        print(f"Connected to {port} at {baud_rate} baud")

        # Wait for the serial connection to initialize
        time.sleep(2)

        # Open the output file
        with open(output_file, 'w') as file:
            start_time = time.time()
            while (time.time() - start_time) < duration:
                if ser.in_waiting > 0:
                    # Read a line from the serial port
                    line = ser.readline().decode('utf-8', errors='ignore').strip()
                    # Print the line to the console
                    print(line)
                    # Write the line to the file
                    file.write(line + '\n')

        # Close the serial connection
        ser.close()
        print("Finished reading serial data.")

    except Exception as e:
        print(f"Error: {e}")

if __name__ == "__main__":
    # Specify the serial port and baud rate
    serial_port = 'COM3'  # Change to your Arduino port, e.g., '/dev/ttyUSB0' on Linux or '/dev/tty.usbmodemXXXX' on macOS
    baud_rate = 9600

    # Duration to read data (in seconds)
    read_duration = 40
    
    # Output file to save the data
    output_file = 'arduino_data.txt'

    # Call the function to read serial data
    read_serial_data(serial_port, baud_rate, read_duration, output_file)
```
